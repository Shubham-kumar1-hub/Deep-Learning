```python
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, LSTM, GRU, Embedding,Bidirectional
from tensorflow.keras.datasets import imdb
```

```python
num_words = 10000    # number of words to keep
(X_train,y_train), (X_test,y_test) = imdb.load_data(num_words = num_words)
```

```python
# Pad Sequences
maxlen=100
X_train = pad_sequences(X_train,maxlen=maxlen, padding = 'post', truncating='post')
X_test = pad_sequences(X_test,maxlen=maxlen, padding = 'post', truncating='post')
```

```python
embedding_dim = 32

model = Sequential([
    Embedding(input_dim=num_words, output_dim=embedding_dim, input_length=maxlen),
    SimpleRNN(5),     # default activation function ->  tanh
    Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

# For Bidirectional 

```python
embedding_dim = 32

model = Sequential([
    Embedding(input_dim=num_words, output_dim=embedding_dim, input_length=maxlen),
    Bidirectional(SimpleRNN(5)),     # default activation function ->  tanh
    Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
embedding_dim = 32

model = Sequential([
    Embedding(input_dim=num_words, output_dim=embedding_dim, input_length=maxlen),
    LSTM(5),     # default activation function ->  tanh
    Dense(1,activation='sigmoid')
])
```
