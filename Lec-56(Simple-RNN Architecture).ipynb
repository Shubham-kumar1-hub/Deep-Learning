```python
from keras import Sequential
from keras.layers import Dense,SimpleRNN
```

```python
model = Sequential()

model.add(SimpleRNN(3,input_shape=(4,5)))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
print(model.get_weights()[0].shape)
model.get_weights()[0]     # between simpleRNN and input
```

```python
print(model.get_weights()[1].shape)
model.get_weights()[1]    # feebback connection
```

```python
print(model.get_weights()[2].shape)
model.get_weights()[2]    # bias of three nodes
```

```python
print(model.get_weights()[3].shape)
model.get_weights()[3]    # output is connected with Simple RNN
```

```python
print(model.get_weights()[4].shape)
model.get_weights()[4]   # bias of output
```
